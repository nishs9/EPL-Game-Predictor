```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="darkgrid")
```

```python
raw_data = pd.read_csv('E0-2018-2019.csv')
raw_data2 = pd.read_csv('E0-2019-2020.csv')
raw_data3 = pd.read_csv('E0-2020-2021.csv')
raw_data = pd.concat([raw_data,raw_data2],axis=0)
raw_data = pd.concat([raw_data,raw_data3],axis=0)
raw_data
```

```python
matches = raw_data[["FTR","HomeTeam","AwayTeam","B365H","B365D","B365A","BWH","BWD","BWA"]]
matches
```

```python
matches.head()
```

```python
plot_size = plt.rcParams["figure.figsize"]
plot_size [0] = 8
plot_size [1] = 6
plt.rcParams["figure.figsize"] = plot_size
```

```python
matches["FTR"].value_counts().plot(kind='pie', autopct='%0.05f%%', colors=['lightblue', 'lightgreen', 'orange'], explode=(0.05, 0.05, 0.05))
```

```python
plt.clf()
home = pd.get_dummies(matches["HomeTeam"],prefix='HomeTeam')
away = pd.get_dummies(matches["AwayTeam"],prefix='AwayTeam')
B365H = matches["B365H"]
B365D = matches["B365D"]
B365A = matches["B365A"]
BWH = matches["BWH"]
BWD = matches["BWD"]
BWA = matches["BWA"]

labels = pd.get_dummies(matches["FTR"],prefix='result')
labels.head()
```

```python
X = pd.concat([home,away,B365H,B365D,B365A,BWH,BWD,BWA],axis=1)
X.head()
```

```python
y = labels.values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from tensorflow.keras.layers import Input, Dense, Activation,Dropout
from tensorflow.keras.models import Model
```

```python
input_layer = Input(shape=(X.shape[1],))
dense_layer_1 = Dense(31, activation='sigmoid')(input_layer)
dense_layer_2 = Dense(15, activation='sigmoid')(dense_layer_1)
output = Dense(y.shape[1], activation='softmax')(dense_layer_2)

model = Model(inputs=input_layer, outputs=output)
opt = tf.keras.optimizers.Adam()
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['acc'])
print(model.summary())

history = model.fit(X_train, y_train, batch_size=8, epochs=100, verbose=2,validation_split=0.2,shuffle=True)
```

```python
plt.plot(history.history["acc"], label="Training Accuracy")
plt.plot(history.history["val_acc"], label="Validation Accuracy")
plt.legend()
```

```python
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
```

```python
score = model.evaluate(X_test, y_test, verbose=2)

print("Test Score:", score[0])
print("Test Accuracy:", score[1])
```
